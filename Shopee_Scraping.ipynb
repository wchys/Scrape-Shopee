```python
#ทำเป็น List
records=[]
data = driver.page_source
soup = bs4.BeautifulSoup(data)
el=soup.select(".row>div")
for e in el: #เพิ่มเนื้อหาต่าง ๆ ที่จะ Scrape ใน Loop นี้
    name = e.select_one("#main > div > div > div > div > div.container.iFOymg > div.doAEtb > div > div.row.shopee-search-item-result__items > div > a > div > div > div.ScPA3O > div.klCFph > div > div").text.strip()
    try:
        p = e.select_one("#main > div > div > div > div > div.container.iFOymg > div.doAEtb > div > div.row.shopee-search-item-result__items > div > a > div > div > div.ScPA3O > div.AQ4KLF > div.cbl0HO.MUmBjS").text.strip()
        if p == '':
            print(1/0)
    except:
        p = e.select_one("#main > div > div > div > div > div.container.iFOymg > div.doAEtb > div > div.row.shopee-search-item-result__items > div > a > div > div > div.ScPA3O > div.AQ4KLF > div").text.strip()
    p = p.replace(",", "").replace("฿", "").replace(" ", "")
    try:
        start_price = float(p)
    except:
        price = p.rpartition('-')[0]
        start_price = float(price)

    sales = e.select_one("#main > div > div > div > div > div.container.iFOymg > div.doAEtb > div > div.row.shopee-search-item-result__items > div > a > div > div > div.ScPA3O > div.tysB0L > div.x\+3B8m.wOebCz").text.strip().replace("พัน","000").replace("ล้าน","000000")
    try:
        sales = float(re.sub('\D','',sales))
    except:
        sales = 0
    province = e.select_one("#main > div > div > div > div > div.container.iFOymg > div.doAEtb > div > div.row.shopee-search-item-result__items > div > a > div > div > div.ScPA3O > div.mrz-bA").text.strip()
    #print(name, p)
    records.append([name, start_price,sales,province])
    
records
```
